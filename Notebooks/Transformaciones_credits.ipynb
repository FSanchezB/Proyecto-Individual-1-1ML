Limpieza y extraccion de **crew**

```python
def filtrar_directores(crew_list):
    directores = [] #lista vacia para guardar a los directores
    for crew in crew_list: #cada crew es un diccionario en la lista de diccionarios
        if crew['job'] == 'Director': # si en la key 'job' el valor es Director, se continua con el ciclo, si no, se va al siguente crew
            directores.append({ #guardamos cada director en la lista
                'id': crew['id'], #'id' crea una nueva key, crew['id'] le dice que le ponga a esa key el valor que tiene ese crew en esa key en el diccionario original
                'job': crew['job'],
                'name': crew['name']
            })
    return directores #regresamos la lista de directores solo con id, job y name
```

```python
df['crew'] = df['crew'].apply(ast.literal_eval) #convertimos los datos a verdaderos diccionarios ya que se guardan como strings
df['crew'] = df['crew'].apply(filtrar_directores) # aplicamos la funcion para transformar y limpiar esa columna
```

```python
def separar_directores(crew_list): #funcion para darle su propia columna
    datos = {} #diccionario vacio para guardar los datos
    for i, director in enumerate(crew_list): #iteramos sobre los directores y le asignamos un indice en i
        datos[f'Director {i+1}'] = director['name'] #creamos una key en el diccionario datos de nombre director y numero indice+1 y le asignamos el valor que contiene director en la key name
        datos[f'Director {i+1} Id'] = director['id']
    return pd.Series(datos) #creamos una serie a partir de datos 
```

```python
df_directores = df['crew'].apply(separar_directores) #creamos un dataframe de directores aplicando la funcion a la columna
```

Limpieza y extraccion de **cast**

```python
def filtrar_cast(cast_list): #funcion para filtar el cast
    cast_filtrado = [] #lista vacia de cast
    for member in cast_list: #iteramos sobre cada diccionario de la lista
        cast_filtrado.append({
            'id': member['id'], #agregamos a cast filtrado una nueva key 'id' y el valor es el id del miembro
            'name': member['name']
        })
    return cast_filtrado #retornamos la lista cast filtrado con solo el id y el nombre
```

```python
df['cast'] = df['cast'].apply(ast.literal_eval) #convertimos los trings a verdaderos diccionarios
df['cast'] = df['cast'].apply(filtrar_cast) #aplicamos la funcion para filtrar los datos del cast
```

```python
df['cast'] #verificamos
```

```python
def separar_cast(cast_list): #funcion para separar
    data = {} #lista vacia 
    for i, member in enumerate(cast_list): #recorremos cada miembro del cast y le asignamos un indice
        data[f'Actor {i+1}'] = member['name'] #creamos la key de actor +1 de su indice y se le asigna el valor de member en el key name
        data[f'Actor {i+1} Id'] = member['id']
    return pd.Series(data) #convertimos en serie de pandas la data
```

```python
cast_df= df['cast'].apply(separar_cast) #hacemos un nuevo df de actores aplicando la funcion a cada columna y la serie resultante seria el key el nombre de la columna y el value el valor
```

```python
id=df['Id_Pelicula']
cast_df.insert(0,'Id_Pelicula',id) #agregamos al inicio una columna con el id
```

```python
def count_movies_by_actor_name(df, actor_name):
    # Select columns containing actor names
    actor_columns = [col for col in df.columns if 'Actor ' in col and 'Id' not in col]

    # Melt the DataFrame to long format
    melted_df = df.melt(id_vars=['Id_Pelicula'], 
                        value_vars=actor_columns, 
                        var_name='actor_column', 
                        value_name='actor_name')

    # Filter by the actor's name
    filtered_df = melted_df[melted_df['actor_name'] == actor_name]

    # Count the number of unique movie IDs
    movie_count = filtered_df['Id_Pelicula'].nunique()

    return movie_count
```
