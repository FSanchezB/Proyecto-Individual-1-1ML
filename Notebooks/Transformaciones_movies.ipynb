```python
import ast
```

```python
def conversion_diccionario(val): #funcion que nos ayuda a convertir los strings pseudo-JSON a diccionarios de python
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError, TypeError):
        return val
```

**Generos**

```python
generos=df_movies['genres'].apply(conversion_diccionario)#convertimos a diccionarios reales
generos=pd.json_normalize(generos) #separamos cada diccionario en una columna distinta
generos
```

```python
g1=generos[0].apply(conversion_diccionario) #convertimos a diccionario
g1=pd.json_normalize(g1) #separamos cada valor del diccionario en una columna nueva
g1.rename(columns={'id':'Id_genero_1','name':'genero_1'},inplace=True) #renombramos
g1.info()
```

```python
g2=generos[1].apply(conversion_diccionario) #lo mismo para genero 2 y 3
g2=pd.json_normalize(g2)
g2.rename(columns={'id':'Id_genero_2','name':'genero_2'},inplace=True)
g2.info()
```

```python
g3=generos[2].apply(conversion_diccionario) #nos quedamos hasta el 3  ya que muy pocas peliculas tienen mas generos e inflariamos el dataset
g3=pd.json_normalize(g3)
g3.rename(columns={'id':'Id_genero_3','name':'genero_3'},inplace=True)
g3.info()
```

```python
generos=pd.concat([g1,g2,g3],axis=1) #df de generos reducido
generos.fillna(0,inplace=True)
generos
```

Production Companies

```python
prod_comp=df_movies['production_companies'].apply(conversion_diccionario)
prod_comp=pd.json_normalize(prod_comp)
prod_comp
```

```python
prod_comp.info() #igual conservaremos 3 companias
```

```python
pc1=prod_comp[0].apply(conversion_diccionario)
pc1=pd.json_normalize(pc1)
pc1.rename(columns={'name':'produccion_1','id':'Id_produccion_1'},inplace=True)
pc1
```

```python
pc2=prod_comp[1].apply(conversion_diccionario)
pc2=pd.json_normalize(pc2)
pc2.rename(columns={'name':'produccion_2','id':'Id_produccion_2'},inplace=True)
pc2
```

```python
pc3=prod_comp[3].apply(conversion_diccionario)
pc3=pd.json_normalize(pc3)
pc3.rename(columns={'name':'produccion_3','id':'Id_produccion_3'},inplace=True)
pc3
```

```python
produccion=pd.concat([pc1,pc2,pc3],axis=1)
produccion.fillna(0,inplace=True)
produccion
```

**Countries**

```python
prod_cont=df_movies['production_countries'].apply(conversion_diccionario)
prod_cont=pd.json_normalize(prod_cont)
prod_cont
```

**idiomas**

```python
spoken=df_movies['spoken_languages'].apply(conversion_diccionario)
spoken=pd.json_normalize(spoken)
spoken
```

```python
spoken.info() #mantendremos 2
```

```python
s1=spoken[0].apply(conversion_diccionario)
s1=pd.json_normalize(s1)
s1.rename(columns={'iso_639_1':'codigo_idioma_1','name':'idioma_1'},inplace=True)
s1
```

**coleccion**

```python
collect=df_movies['belongs_to_collection'].apply(conversion_diccionario)
collect=pd.json_normalize(collect)
collect #vemos las columnas poster_path y backdrop_path y no sirven para recomendar, asi que las vamos a droppear
```
