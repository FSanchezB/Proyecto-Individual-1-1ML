```python
import ast
```

```python
def conversion_diccionario(val): #funcion que nos ayuda a convertir los strings pseudo-JSON a diccionarios de python
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError, TypeError):
        return val
```

```python
generos=df_movies['genres'].apply(conversion_diccionario)#convertimos a diccionarios reales
generos=pd.json_normalize(generos) #separamos cada diccionario en una columna distinta
generos
```

```python
prod_comp=df_movies['production_companies'].apply(conversion_diccionario)
prod_comp=pd.json_normalize(prod_comp)
prod_comp.info()
```

```python
prod_cont=df_movies['production_countries'].apply(conversion_diccionario)
prod_cont=pd.json_normalize(prod_cont)
prod_cont.info()
```

```python
spoken=df_movies['spoken_languages'].apply(conversion_diccionario)
spoken=pd.json_normalize(spoken)
spoken.info()
```

```python
collect=df_movies['belongs_to_collection'].apply(conversion_diccionario)
collect=pd.json_normalize(collect)
collect #vemos las columnas poster_path y backdrop_path y no sirven para recomendar, asi que las vamos a droppear
```

```python
collect.drop(['poster_path','backdrop_path'],axis=1,inplace=True) #droppeamos y verificamos
collect
```
